# PR√â-PROCESSAMENTO DOS POSTS EM PORTUGU√äS DO BRASIL

```python
# Carregar arquivo
df = pd.read_json('./totalData/darkweb/deepPtbr.json')
```

```python
# Numero de posts/perguntas
num_linhas = df['id'].count()
num_linhas
```

```python
# Filtrar os dados a partir da data '10/09/2022' ('2022-09-10')
df['dataCreated'] = pd.to_datetime(df.dataCreated, format='%Y-%m-%d %H:%M:%S')
df['dataCreated'] = df['dataCreated'].dt.tz_localize(None)

#df = df[df['dataCreated'] >= pd.to_datetime('2022-07-15')]
```

```python
# Selecionar colunas
df = df.loc[:,['id','category','title','question','answers','tags','dataCreated', 'year', 'quarter']]

# Refazer o √≠ndice
df = df.reset_index(drop=True)
df = df.sort_values('id')
```

```python
df
```

```python
import re
import nltk
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from unidecode import unidecode


###############

df['title'] = df['title'].astype(str)
df['question'] = df['question'].astype(str)
df['answers'] = df['answers'].astype(str)

full_text_col = 'full_text'
required_columns = ["title", "question", "answers", full_text_col]
df[full_text_col] = df["title"] + " " + df["question"] + " " + df["answers"]

###############

nltk.download('stopwords')

stop_words = set(stopwords.words('portuguese'))
stop_words.update(["pra","etc", "none", "vai", "ter", "nan", "user", "author", "title", "none",
                    "name", "score", "content", "down", "votes", "created", "comments", "comment",
                    "answercontent", "vote", "type", "points", "aqui", "pode", "sobre", "fazer",
                    "alguem", "tudo", "regular", "coisa", "bem", "vou", "sei", "boca", "algum",
                    "alguns", "alguma", "algo", "nada", "bom", "entao", "acho", "quer", "the",
                    "and", "you", "cara", "coisas", "sim", "ainda", "ver", "usar", "assim",
                    "index" , "upvotes", "questionid", "answerid", "downvotes", "datacreated", 
                    "answers", "answercontent"])

# Substituir caracteres com acentos por seus equivalentes sem acentos
stop_words_sem_acentos = set()
for palavra in stop_words:
    stop_words_sem_acentos.add(unidecode(palavra))
stop_words = stop_words_sem_acentos


def preprocess_text(text):    
    # Substituir caracteres com acentos por seus equivalentes sem acentos
    text = unidecode(text)
    # Converter o texto para min√∫sculo
    text = text.lower()
    # Substituir \\n por um espa√ßo em branco
    text = re.sub(r'\s*\\n\s*', ' ', text)
    # Remover tags HTML
    text = BeautifulSoup(text, 'html.parser').get_text()
    # Remover links .onion
    text = re.sub(r'\S*\.onion\S*', ' ', text)
    # Remover URLs
    text = re.sub(r'http\S+', ' ', text)
    # Remover os title que aparecem em answers
    text = re.sub(r"'title': [^,]*,", ' ', text)
    # Remover os nomes de usu√°rio que aparece em answers
    text = re.sub(r"'name': [^,]*,", ' ', text)
    # Remover os type que aparecem em answers 
    text = re.sub(r"'type': [^,]*,", ' ', text)
    # Remover os author que aparecem em answers 
    text = re.sub(r"'author': [^,]*,", ' ', text)
    # Remover sequ√™ncia de kkkk
    text = re.sub(r'k{2,}\S*', ' ', text)
    
    # Remover termos com mais de 4 consoantes seguidas
    consoantes_5m = "([bcdfghjklmnpqrstvwxyz]{5,})"
    text = re.sub(consoantes_5m, " ", text)
    # Remover termos com mais de 5 vogais seguidas
    vogais_6m = "([aeiou]{6,})"
    text = re.sub(vogais_6m, " ", text)
    
    # Substituir caracteres que n√£o s√£o letras por espa√ßos em branco
    text = re.sub('[^A-Za-z]+', ' ', text)
    # Remover espa√ßos em branco extras e substituir sequ√™ncias de espa√ßos em branco por um √∫nico espa√ßo em branco
    text = re.sub('\s+', ' ', text.strip())
    # Remover stop words
    words = text.split()
    words = [word for word in words if word not in stop_words]
    words = [word for word in words if len(word) > 2]
    text = ' '.join(words)
    return text
```

```python
# Aplicar a fun√ß√£o de pr√©-processamento √† df[full_text_col]
df[full_text_col] = df[full_text_col].apply(preprocess_text)
```

```python
# Selecionar colunas
# df = df.loc[:,['id', 'category', 'title', 'author', 'question', 'answers', 'dataCreated', 'year', 'quarter']]
df = df.loc[:,['id', 'forum', 'title', 'user', 'post', 'commentaries', 'dataCreated']]

# Refazer o √≠ndice
df = df.reset_index(drop=True)
df = df.sort_values('id')
```

```python
df['dataCreated'] = pd.to_datetime(df.dataCreated, format='%Y-%m-%d %H:%M:%S')
df['dataCreated'] = df['dataCreated'].dt.tz_localize(None)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from unidecode import unidecode
from bs4 import BeautifulSoup

###############

df['title'] = df['title'].astype(str)
#df['question'] = df['question'].astype(str)
#df['answers'] = df['answers'].astype(str)
df['post'] = df['post'].astype(str)
df['commentaries'] = df['commentaries'].astype(str)

full_text_col = 'full_text_ingles'
#required_columns = ["title", "question", "answers", full_text_col]
#df[full_text_col] = df["title"] + " " + df["question"] + " " + df["answers"]
required_columns = ["title", "post", "commentaries", full_text_col]
df[full_text_col] = df["title"] + " " + df["post"] + " " + df["commentaries"]

nltk.download('stopwords')

##############

stop_words_eng = set(stopwords.words('english'))
stop_words_eng.update(["index" , "upvotes", "postid", "comentaryid", "downvotes", "datacreated", "answers",
                       "questionid", "answerid", "content" , "comments", "answercontent", "answered",
                       "author", "question", "points", "commented", "comment", "commentary", "id", "votes"
                       "user", "jpg", "nan", "none", "www", "https"])


def preprocess_text(text):    
    # Substituir caracteres com acentos por seus equivalentes sem acentos
    text = unidecode(text)
    # Converter o texto para min√∫sculo
    text = text.lower()
    # Remover tags HTML
    text = BeautifulSoup(text, 'html.parser').get_text()
    # Remove pontua√ß√µes e caracteres especiais
    text = re.sub(r'[^\w\s]', '', text)
    # Substituir caracteres que n√£o s√£o letras por espa√ßos em branco
    text = re.sub('[^A-Za-z]+', ' ', text)
    # Remover espa√ßos em branco extras e substituir sequ√™ncias de espa√ßos em branco por um √∫nico espa√ßo em branco
    text = re.sub('\s+', ' ', text.strip())
    # Remover stop words
    words = text.split()
    words = [word for word in words if word not in stop_words_eng]
    text = ' '.join(words)
    return text
```

```python
# Aplicar a fun√ß√£o de pr√©-processamento √† df[full_text_ingles]
df['full_text_ingles'] = df['full_text_ingles'].apply(preprocess_text)
```
